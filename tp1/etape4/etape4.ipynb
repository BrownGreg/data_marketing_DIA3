### Hypothèse 4 -

Je suppose que les clients situés au Royaume-Uni, qui est le pays comptant le plus grand nombre de clients, sont des clients plus anciens. En effet, leur panier moyen ne figure pas dans le top 10, alors que nous avons observé une augmentation du panier moyen chez les clients récents.
